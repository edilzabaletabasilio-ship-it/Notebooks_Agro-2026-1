# **Control de Rendimiento en Extracción de Aceite**
Este ejercicio es ideal para introducir float, int y la concatenación de strings.

**Planteamiento:**

Una planta procesadora de palma de aceite necesita calcular el **Índice de Rendimiento de Extracción (IRE)**. El operario debe ingresar el peso total de los racimos de fruta fresca (RFF) procesados en kilogramos y la cantidad de aceite crudo obtenido en litros.

**Requerimientos:**

1. Solicitar el nombre del lote de producción (Tipo string).
2. Solicitar el peso de los racimos en kg (Tipo float).(Pr)
3. Solicitar los litros de aceite obtenidos (Tipo float).(lt)
4. alcular el rendimiento sabiendo que la densidad aproximada del aceite es de 0.92 kg/L. (r)
$$Pa (lt)=Litros*0.92$$


$$r=(Pa/Pr)*100$$

**Prueba del programa:**

Si se ingresan 1000 kg de racimos y se obtienen 200 litros de aceite, el programa debe informar que el rendimiento del lote es del 18.4%.

````python
##ENTRADA## Solicitar el nombre del lote de producción (Tipo string).
##Solicitar el peso de los racimos en kg (Tipo float).(Pr)##
##SOLICITAR## Solicitar los litros de aceite obtenidos (Tipo float).(lt)
Nombre = input("Ingrese el nombre del lote: ")
Pr = float(input("Ingrese el peso de los racimos en kg: "))
lt = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))

# POCESO calcula el peso del aceite (pa) y el rendimiento (r)
pa = lt * 0.92
r = (pa / Pr) * 100

#SALIDA informar el rendimiento del lote
print("El rendimiento del lote", Nombre, "es del", r, "%")
````

```markdown
## Explicación del Código de Control de Rendimiento de Aceite

Este código solicita al usuario información sobre un lote de producción de aceite de palma, realiza cálculos para determinar el rendimiento y luego muestra el resultado.

### Entrada de Datos:

*   `Nombre = input("Ingrese el nombre del lote: ")`:
    *   Esta línea solicita al usuario que ingrese el nombre del lote de producción. El texto entre comillas es el mensaje que se muestra al usuario. La entrada del usuario se almacena como una cadena de texto (string) en la variable `Nombre`.

*   `Pr = float(input("Ingrese el peso de los racimos en kg: "))`:
    *   Aquí se pide al usuario que ingrese el peso total de los racimos de fruta fresca (RFF) procesados en kilogramos. La función `input()` recoge la entrada como texto, y `float()` la convierte a un número decimal (float) porque el peso puede tener decimales. El valor se guarda en la variable `Pr`.

*   `lt = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))`:

### Procesamiento de Datos (Cálculos):

*   `pa = lt * 0.92`:

### Salida de Resultados:

*   `print("El rendimiento del lote", Nombre, "es del", r, "%")`:
    *   Finalmente, esta línea muestra el resultado al usuario. Utiliza la función `print()` para concatenar varias cadenas de texto y variables, formando un mensaje legible que informa sobre el rendimiento del lote específico (`Nombre`) y su porcentaje (`r`).
```

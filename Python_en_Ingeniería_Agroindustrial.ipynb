## ¿Qué es Python?

A grandes rasgos, Python es un lenguaje de programación de **alto nivel**, lo que significa que su sintaxis es muy parecida al lenguaje humano (específicamente al inglés). Es famoso por ser:

## 3 Ejemplos Sencillos para Agroindustria

### 1. Cálculo del Índice de Cosecha

Este código ayuda a determinar si un fruto está listo basándose en su peso fresco y peso seco.

```
# Cálculo del porcentaje de materia seca
peso_fresco = 150.5  # gramos
peso_seco = 22.3     # gramos

porcentaje_ms = (peso_seco / peso_fresco) * 100

print(f"El porcentaje de materia seca es: {porcentaje_ms:.2f}%")

if porcentaje_ms > 14:
    print("Estado: Óptimo para procesamiento.")
else:
    print("Estado: Falta maduración.")
```

### 2. Conversión de Escalas de Temperatura

Fundamental para procesos de pasteurización o deshidratación donde los equipos vienen en distintas unidades.

```
def celsius_a_fahrenheit(c):
    return (c * 9/5) + 32

temp_proceso = 72  # Grados Celsius para pasteurización rápida
temp_f = celsius_a_fahrenheit(temp_proceso)

print(f"La temperatura de pasteurización es {temp_proceso}°C ({temp_f}°F)")
```

### 3. Estimación de Capacidad de Almacenamiento

Calcula cuántas toneladas de grano caben en un silo cilíndrico.

```
import math

radio = 3.5    # metros
altura = 10.0  # metros
densidad_maiz = 0.75 # toneladas por metro cúbico (aprox)

volumen = math.pi * (radio**2) * altura
capacidad_total = volumen * densidad_maiz

print(f"El silo tiene una capacidad de: {capacidad_total:.2f} toneladas de maíz.")
```
